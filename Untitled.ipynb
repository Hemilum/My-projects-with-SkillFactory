```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations

data = pd.read_csv('movie_bd_v5.csv')

answers = {}

# расчет прибыли по фильму и добавление колонки с прибылью в таблицу.
budget = data.budget
revenue = data.revenue
profit = revenue - budget
data['profit'] = profit


# с колонкой дата в том виде, что есть, работать не удобно
# исходя из вопросов, я решил добавить две новых колонки
# одна с годом и другая с месяцем

def year(release_date):
    year = release_date[-4:]
    return int(year)

def month(date):
    slash = date.find('/')
    month = date[ : slash]
    return int(month)

month = data.release_date.apply(month)
year = data.release_date.apply(year)

data['month'] = month
data['year'] = year

data['film_name_len'] = data.original_title.apply(len)

```
